```python
import numpy as np
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
import datetime
import csv
import os
from sklearn.metrics import r2_score, mean_squared_error

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
from modules.prediction import load_all_data
from modules.prediction import precrime_train_test_split
from modules.prediction import load_splits
from modules.prediction import create_all_splits
from modules.prediction import sample_model
from skopt import gp_minimize
from modules.gb_reg import GBR, gbr_model, gbr_model_bo
```

```python
crime_data = load_all_data()
splits = load_splits()
train_test_data = create_all_splits(crime_data, splits)
```

```python
X_train_fine, X_test_fine, y_train_fine, y_test_fine = train_test_data['fine']
X_train_coarse, X_test_coarse, y_train_coarse, y_test_coarse = train_test_data['coarse']
X_train_2016, X_test_2016, y_train_2016, y_test_2016 = train_test_data['2016']
```

```python
parameters_coarse=gbr_model_bo(X_train_coarse.head(1000), y_train_coarse.head(1000), X_test_coarse.head(100))
```

```python
parameters_coarse
```

```python
y_pred_coarse=gbr_model(X_train_coarse, y_train_coarse, X_test_coarse, parameters_coarse)
```

```python
for crime_type in y_train_coarse.select_dtypes(exclude=['object']).columns:
    print('{0}: R2 = {1:.1f}, MSE = {2:.4f}'.format(
            crime_type,
            100*r2_score(y_test_coarse[crime_type], y_pred_coarse[crime_type]),
            mean_squared_error(y_test_coarse[crime_type], y_pred_coarse[crime_type]
    )))
```

```python
os.makedirs('../model_predictions/coarse/', exist_ok=True)
y_pred_coarse.to_csv('../model_predictions/coarse/gbr.csv', quoting=csv.QUOTE_NONNUMERIC)
```
